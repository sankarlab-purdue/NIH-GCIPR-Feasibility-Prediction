```python
import sys
sys.path.append('../core')
from FeasibilityAnalyzer import *
from utils import *
```

```python
# RDenzyme RetroSim files
rdenzyme_file = '../data/rdenzyme_db.pkl'
retrobiocat_file = '../data/retrobiocat_db.pkl'

# Directionality dataset for feasibility assessment
directionality_file = '../data/directionality_dataset.pkl'

# buyables and custom datasets
buyable_file = '../data/sampled_split.db'
custom_file = '../data/custom.db'
buyables = sqlite3.connect(f'file:{buyable_file}', uri=True).cursor()
custom = sqlite3.connect(f'file:{custom_file}', uri=True).cursor()

# Load retrosim
retro_analyzer = Retrosim(rdenzyme_file, retrobiocat_file)
feas_analyzer = FeasibilityAnalyzer(directionality_file)
```

```python
# User input
target = "C#C[C@]1([C@H](C[C@@H](O1)N2C=NC3=C(N=C(N=C32)F)N)O)CO"
max_precursors = 15
retrobiocat = False

# Run single-step retrosynthesis analysis
outcomes = retro_analyzer.single_step_retro(target, max_precursors, False, retrobiocat)
```

#### Check results without feasibility assessment

```python
# Display without assessing feasibility
display_single_step_results(target, outcomes, buyables, custom)
```

#### Check results with feasibility assessment

```python
# Check feasibility of proposed steps
feasible, not_feasible, unknown = feas_analyzer.single_step_feasibility(outcomes, target)
```

```python
# Display feasible results
display_single_step_results(target, feasible, buyables, custom)
```

```python
# Display not feasible results
display_single_step_results(target, not_feasible, buyables, custom)
```

```python
# Display results with unknown feasibility - feasibility assessment was not carried out
display_single_step_results(target, unknown, buyables, custom)
```
