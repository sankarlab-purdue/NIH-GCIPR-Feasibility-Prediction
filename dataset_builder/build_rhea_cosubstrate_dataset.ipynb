```python
cofactor_list = pd.read_pickle("acquired_data/Cofactor_list.pkl")
rxn_smiles_data = pd.read_csv("acquired_data/06182025_rhea-reaction-smiles.tsv", 
                         delimiter="\t", names=['RHEA_ID', 'rxn_smiles'], engine='python')
```

```python
cofactor_by_id = {
    'id': [],
    'cofactors_left': [],
    'cofactors_right': []
}    

for idx, row in rxn_smiles_data.iterrows():
    _id = row['RHEA_ID']
    rxn_smiles = canonicalize_reaction(row['rxn_smiles'])

    cofactor_by_id['id'].append(_id)

    cft_left = []
    cft_right = []
    rct, prd = rxn_smiles.split(">>")
    for item in rct.split('.'):
        if item in cofactor_list:
            cft_left.append(item)

    if cft_left:
        cofactor_by_id['cofactors_left'].append('.'.join(cft_left))
    else:
        cofactor_by_id['cofactors_left'].append("nan")

    for item in prd.split('.'):
        if item in cofactor_list:
            cft_right.append(item)

    if cft_right:
        cofactor_by_id['cofactors_right'].append('.'.join(cft_right))
    else:
        cofactor_by_id['cofactors_right'].append("nan")
```

```python
def add_missing_components(rhea_id=None, precursors=None, target=None, cofactors=None, debug=False):

    matching_row = cofactors[cofactors['id'] == rhea_id]
    cofactors_right = matching_row['cofactors_right'].iloc[0]
    cofactors_left = matching_row['cofactors_left'].iloc[0]
    
    # Build left side (from precursors)
    left_molecules = precursors
    if not pd.isna(cofactors_left):
        for cofactor in cofactors_left.split('.'):
            if cofactor not in precursors and cofactor != "nan":
                left_molecules.append(cofactor)
    
    
    # Build right side (from target)
    right_molecules = [target]
    if not pd.isna(cofactors_right):
        for cofactor in cofactors_right.split('.'):
            if cofactor != target and cofactor != "nan":
                 right_molecules.append(cofactor)
                    
    # fully assembled reaction
    full_reaction = '.'.join(left_molecules) + '>>' + '.'.join(right_molecules)
    
    if debug:
        print(full_reaction)

    return full_reaction
```

```python
# First, check for duplicates
already_seen = []
duplicate = []

for idx, row in rdenzyme_file.iterrows():
    _id = str(row['id'])

    if _id in already_seen:
        duplicate.append(_id)
    else:
        already_seen.append(_id)
```

```python
import rdkit.Chem as Chem

def remove_atom_map(smi):

    react, prod = smi.split('>>')

    react_mol = Chem.MolFromSmiles(react)
    prod_mol = Chem.MolFromSmiles(prod)

    for atom in react_mol.GetAtoms():
        atom.SetAtomMapNum(0)

    for atom in prod_mol.GetAtoms():
        atom.SetAtomMapNum(0)

    return Chem.MolToSmiles(react_mol)+'>>'+Chem.MolToSmiles(prod_mol)
```

```python
missing_components = {
    'id': [],
    'missing_comp_left': [],
    'missing_comp_right': []
}

duplicate_counters = {}

for idx, row in rdenzyme_file.iterrows():
    _id = row['id']
    template_reaction = canonicalize_reaction(remove_atom_map(row['rxn_smiles']))
    full_rhea_reaction = canonicalize_reaction(row['not atom mapped smiles-input'])

    str_id = str(_id)
    if str_id in duplicate:
        count = duplicate_counters.get(str_id, 0)
        missing_components['id'].append(f"{str_id}-{count}")
        duplicate_counters[str_id] = count + 1
    else:
        missing_components['id'].append(str_id)

    rct_temp, prd_temp = template_reaction.split(">>")
    rct_full, prd_full = full_rhea_reaction.split(">>")

    rct_temp_list = rct_temp.split('.')
    prd_temp_list = prd_temp.split('.')

    comp_left = [item for item in rct_full.split('.') if item not in rct_temp_list]
    missing_components['missing_comp_left'].append('.'.join(comp_left) if comp_left else "nan")

    comp_right = [item for item in prd_full.split('.') if item not in prd_temp_list]
    missing_components['missing_comp_right'].append('.'.join(comp_right) if comp_right else "nan")
```

```python
def add_missing_components(rhea_id=None, precursors=None, df=None, target=None, debug=False):

    matching_row = df[df['id'] == str(rhea_id)]

    # Check if matching_row is empty
    if matching_row.empty:
        if debug:
            print(f"Warning: No cofactor data found for RHEA ID {rhea_id}")
        # Return a basic reaction without cofactors
        return '.'.join(precursors) + '>>' + target
        
    cofactors_right = matching_row['missing_comp_right'].iloc[0]
    cofactors_left = matching_row['missing_comp_left'].iloc[0]

    # Build left side (from precursors)
    left_molecules = precursors
    if not pd.isna(cofactors_left):
        for cofactor in cofactors_left.split('.'):
            if cofactor not in precursors and cofactor != "nan":
                left_molecules.append(cofactor)
                
    # Build right side (from target)
    right_molecules = [target]
    if not pd.isna(cofactors_right):
        for cofactor in cofactors_right.split('.'):
            if cofactor != target and cofactor != "nan":
                    right_molecules.append(cofactor)
        
    # fully assembled reaction
    full_reaction = '.'.join(left_molecules) + '>>' + '.'.join(right_molecules)

    if debug:
        print(f"Adding missing reaction components: {full_reaction}")

    return full_reaction
```
